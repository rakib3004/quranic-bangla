# Fetch Sura Doc when hits into
This url : [Surah Fatiha](https://www.hadithbd.com/quran/byword/detail/?sura=1)

```python
!pip install os
```

```python
import os
import requests
import re

# Replace this URL with the actual URL you want to fetch files from
url = "https://www.hadithbd.com/quran/byword/detail/?sura=1"

response = requests.get(url)

if response.status_code == 200:
    # Create a directory to save the files if it doesn't exist
    if not os.path.exists("src/downloaded_files"):
        os.makedirs("src/downloaded_files")

    # Extract filenames from the content-disposition header
    content_disposition = response.headers.get("content-disposition")
    filenames = re.findall(r'filename="(.*?)"', content_disposition) if content_disposition else []

    for filename in filenames:
          file_url = url + filename
          file_response = requests.get(file_url)

          if file_response.status_code == 200:
              # Save the file to the specified directory
              file_path = os.path.join("downloaded_files", filename)
              with open(file_path, "wb") as file:
                  file.write(file_response.content)

              print(f"Downloaded and saved: {filename}")
          else:
              print(f"Failed to download: {filename}")
else:
    print("Failed to fetch files")

print("All doc files downloaded and saved.")
```

```python
import os
import re
import pandas as pd
from bs4 import BeautifulSoup

# Function to extract verse information from a surah file
def extract_verse_info(surah_file):
    with open(surah_file, 'r', encoding='utf-8') as file:
        content = file.read()

    soup = BeautifulSoup(content, 'html.parser')


    verses = []
    for verse_tag in soup.find_all('p', class_='text-info text-right'):
        verse_number = (verse_tag.text)
        arabic_word = verse_tag.find_next('p', dir='rtl').text
        english_word = verse_tag.find_next('p', dir='ltr', class_='font-kalpurush-reading').text
        bangla_word = verse_tag.find_next('p', dir='ltr', class_='font-kalpurush-reading').text

        verses.append({
            'Verse Number': verse_number,
            'Arabic Word': arabic_word,
            'Bangla Word': bangla_word
        })

    return verses

# Path to the directory containing surah files
surah_dir = '/content/drive/MyDrive/SPL3/Colab/'

# List to store verse information from all surahs
all_verses = []

# Iterate through surah files and extract verse information
for filename in os.listdir(surah_dir):
    if filename.endswith('.html'):
        surah_file = os.path.join(surah_dir, filename)
        verses = extract_verse_info(surah_file)
        all_verses.extend(verses)

# Create a pandas DataFrame
quran_df = pd.DataFrame(all_verses)

# Display the DataFrame
# pd.set_option('display.max_rows', None)
# print(quran_df)

output_csv_path = os.path.join(surah_dir, 'quran_data.csv')
quran_df.to_csv(output_csv_path, index=False)

print(f"Quran data saved to: {output_csv_path}")
```
